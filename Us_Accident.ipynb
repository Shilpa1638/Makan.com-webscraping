Start_Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
plt.figure(figsize=(15,10))
sns.histplot(data = df.Start_Time.dt.hour, stat = 'density', bins=24, color = 'seagreen')
plt.xticks(fontsize=15)

plt.yticks(fontsize=15)
plt.xlabel('24 hours', fontsize=20)
plt.ylabel('Density of accidents', fontsize=20)

plt.title('Distribution of accidents over a period of 24 hours ', fontsize = 20)

plt.show()
```

From above graph we can see that higher number of accident occures between 6 Am to 10 Am in morning and 2 Pm to 6 Pm.

Distribution of Accident over the week

```python
plt.figure(figsize=(10,5))
sns.histplot(data = df.Start_Time.dt.dayofweek,stat = 'density', bins=7, color = 'seagreen')
plt.xticks(fontsize=15)

plt.yticks(fontsize=15)
plt.xlabel('Days of week', fontsize=20)
plt.ylabel('Density of accidents', fontsize=20)

plt.title('Distribution of accidents over a week ', fontsize = 20)

plt.show()
```

We can observe from the plot that the distribution of accidents over weekdays almost remain constant while it decreases for weekends probably because of holidays. (0 = Monday, 1 = Tuesday, 2 = Wednesday, 3 = Thursday, 4 = Friday, 5 = Saturday, 6 = Sunday)



#Temperature

```python
df['Temperature(F)']
```

Lets segregate Accidents by Hot and Cold climate.

```python
cold_temp_count = len(df[df['Temperature(F)'] < 70.0])
hot_temp_count = len(df[df['Temperature(F)'] >= 70.0])
```

```python
lab = ['Accidents in cold climate', 'Accidents in hot climate', 'Accidents without temperature records']
```

```python
plt.figure(figsize=(20,10))
plt.pie([cold_temp_count, hot_temp_count, len(df) - cold_temp_count - hot_temp_count], labels = lab, autopct = '%0.2f%%', shadow = True, explode = [0.2,0.2,0.2], textprops={'fontsize': 10})
plt.title('Accident Percentages in cold and hot climate ', fontsize = 20)
plt.show()
```

from above pie chart we can see that 62 % of accidents occur in cold climate and 35 % accidents occure in hot climate.
